```python
import pandas as pd
from pymongo import MongoClient
from sqlalchemy import create_engine
```

```python
def get_video_channel_names():
    # Create the database connection
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    
    # Define the query
    query = '''SELECT channel.channel_name, video.video_name 
    FROM channel 
    JOIN playlist 
    JOIN video ON channel.channel_id = playlist.channel_id AND playlist.playlist_id = video.playlist_id
    '''
    # Execute the query and fetch the results
    result = pd.read_sql_query(query, engine)
    
    # Close the database connection
    engine.dispose()
    
    return result
video_channel_names = get_video_channel_names()
video_channel_names
```

```python
def get_top_10_most_viewed_videos():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = '''
    SELECT channel.channel_name, video.video_name, video.view_count 
    FROM channel 
    JOIN playlist ON channel.channel_id = playlist.channel_id 
    JOIN video ON playlist.playlist_id = video.playlist_id 
    ORDER BY video.view_count DESC LIMIT 10
    '''
    result = pd.read_sql_query(query, engine)
    return result


# Get top 10 most viewed videos and their respective channels
top_10_most_viewed_videos = get_top_10_most_viewed_videos()
top_10_most_viewed_videos
```

```python
def get_videos_with_highest_likes():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = '''
    SELECT channel.channel_name, video.video_name, video.like_count 
    FROM channel 
    JOIN playlist ON channel.channel_id = playlist.channel_id 
    JOIN video ON playlist.playlist_id = video.playlist_id 
    ORDER BY video.like_count DESC
    '''
    result = pd.read_sql_query(query, engine)
    return result

# Get videos with the highest number of likes
videos_highest_likes = get_videos_with_highest_likes()
videos_highest_likes
```

```python
def get_total_likes_per_video():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = '''
    SELECT video.video_name,
    SUM(video.like_count) AS total_likes FROM video
    GROUP BY video.video_name
    '''
    result = pd.read_sql_query(query, engine)
    return result

# Get total likes and dislikes per video
total_likes_per_video = get_total_likes_per_video()
total_likes_per_video
```

```python
def get_total_views_per_channel():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = '''
    SELECT channel.channel_name, SUM(video.view_count) AS total_views
    FROM channel
    JOIN playlist ON channel.channel_id = playlist.channel_id
    JOIN video ON playlist.playlist_id = video.playlist_id
    GROUP BY channel.channel_name
    ORDER BY total_views DESC;
    '''
    result = pd.read_sql_query(query, engine)
    return result

# Get total views per channel
total_views_per_channel = get_total_views_per_channel()
total_views_per_channel
```

```python
def get_channels_published_in_year():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = f'''
    SELECT channel.channel_name, video.video_name, video.published_date 
    FROM channel JOIN playlist ON channel.channel_id = playlist.channel_id 
    JOIN video ON playlist.playlist_id = video.playlist_id 
    WHERE EXTRACT(YEAR FROM published_date) = 2022
    '''
    result = pd.read_sql_query(query, engine)
    return result


# Get names of channels published in the year 2022
channels_published_in_2022 = get_channels_published_in_year()
channels_published_in_2022
```

```python
def get_average_duration_per_channel():
    engine = create_engine('mysql+pymysql://root:Ajith568.@localhost:3306/youtube_data')
    query = '''
    SELECT channel.channel_name, 
    SEC_TO_TIME(AVG(TIME_TO_SEC(video.duration))) AS duration
    FROM channel
    JOIN playlist ON channel.channel_id = playlist.channel_id
    JOIN video ON playlist.playlist_id = video.playlist_id
    GROUP BY channel_name
    ORDER BY duration DESC
    '''
    result = pd.read_sql_query(query, engine)
    return result

# Get average duration of videos per channel
average_duration_per_channel = get_average_duration_per_channel()
average_duration_per_channel
```
